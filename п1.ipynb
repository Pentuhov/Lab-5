```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1980 + rng
 
fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(yrs, rng + rnd, labels=['Саудовская Аравия', 'Катар', 'Китайская Республика'])
ax.set_title('Внешний долг долар')
ax.legend(loc='upper left')
ax.set_ylabel('Общая задолженность')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
 
plt.show()
```
